```python
import pandas as pd
import glob
import os

# Load all CSVs
files = glob.glob("*_clean.csv")
files.sort()  # Ensure chronological order

df_list = []
for f in files:
    temp = pd.read_csv(f)
    # Extract date from filename
    date_str = f.split("/")[-1].split("_")[0]
    temp["file_date"] = date_str
    df_list.append(temp)

# Combine
combined_df = pd.concat(df_list, ignore_index=True)

# Get last file name for naming convention
last_file = os.path.basename(files[-1])
last_name = last_file.replace("_clean.csv", "")

# Define output filename
output_file = f"df_upto_{last_name}.csv"

# Remove old combined files if they exist
for old in glob.glob("df_upto_*.csv"):
    os.remove(old)

# Save new combined file
combined_df.to_csv(output_file, index=False)
print(f"Saved combined dataset as {output_file}")
```

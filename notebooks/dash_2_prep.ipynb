```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# --- Step 1: Compute precision deviation per school ---
school_precision = (
    filtered_df.groupby("School")
    .apply(lambda g: g["Actual_Meals_Served"].sum() / g["Projected_Kids"].sum() - 1)
    .reset_index(name="Precision_Deviation")
)

# --- Step 2: Sort by absolute deviation (closest to 0 = best precision) ---
school_precision["AbsDeviation"] = school_precision["Precision_Deviation"].abs()
top5_schools = school_precision.sort_values("AbsDeviation").head(5)

# --- Step 3: Plot using Plotly Graph Objects ---
fig = go.Figure()

fig.add_trace(go.Bar(
    x=top5_schools["School"],
    y=top5_schools["Precision_Deviation"],
    text=top5_schools["Precision_Deviation"].round(3),
    textposition="outside",
    marker_color="skyblue"
))

# Add reference line at 0 (perfect precision)
fig.add_hline(y=0, line_dash="dash", line_color="red")

# Layout adjustments
fig.update_layout(
    title="Top 5 Schools with Best Estimation Precision (Deviation Form)",
    xaxis_title="School",
    yaxis_title="Deviation (Actual/Projected - 1)",
    yaxis=dict(
        zeroline=True,
        zerolinecolor="red",
        zerolinewidth=2,
    
    )
)

fig.show()


```

```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster, MiniMap

# Load data
dmap = pd.read_csv("https://raw.githubusercontent.com/samy-migwi/Githunguri-kitchen-f4e-dashboard/main/data/map2.csv")
dmap.columns = dmap.columns.str.strip().str.replace(' ', '_').str.replace('__', '_')

def generate_folium_map(selected_school=None):
    # Color map for divisions
    division_colors = {
        'Githunguri i': '#3B82F6',      # Blue
        'Githunguri II': '#22C55E',     # Green  
        'Githunguri I': '#3B82F6',      # Blue (alternate spelling)
        'Cluster 12': '#A855F7',        # Purple
    }
    
    # Clean division data
    dmap['Division'] = dmap['Division'].str.strip()
    
    # Calculate bounds
    min_lat = dmap['lat'].min() - 0.02
    max_lat = dmap['lat'].max() + 0.02
    min_lon = dmap['lon'].min() - 0.02
    max_lon = dmap['lon'].max() + 0.02
    
    center_lat = dmap['lat'].mean()
    center_lon = dmap['lon'].mean()
    
    # Create map with multiple tile options
    m = folium.Map(
        location=[center_lat, center_lon],
        zoom_start=13,
        tiles=None,  # We'll add custom tiles
        control_scale=True,
    )
    
    # Add tile layers
    folium.TileLayer(
        'CartoDB positron',
        name='Clean',
        control=True,
        opacity=0.8
    ).add_to(m)
    
    folium.TileLayer(
        'OpenStreetMap',
        name='Detailed',
        control=True
    ).add_to(m)
    
    folium.TileLayer(
        'CartoDB dark_matter',
        name='Dark',
        control=True
    ).add_to(m)
    
    # Add satellite option using ESRI
    folium.TileLayer(
        tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
        attr='Esri',
        name='Satellite',
        control=True,
        overlay=False
    ).add_to(m)
    
    # Feature groups for layer control
    divisions_group = folium.FeatureGroup(name='Schools by Division', control=False)
    names_group = folium.FeatureGroup(name='School Names', control=True)
    coverage_group = folium.FeatureGroup(name='Coverage Areas', control=True)
    
    # Marker cluster for better performance with many schools
    marker_cluster = MarkerCluster(name='All Schools').add_to(m)
    
    # Column names
    school_col = 'School'
    lat_col = 'lat'
    lon_col = 'lon'
    division_col = 'Division'
    f4e_col = 'F4E_Sponsored' if 'F4E_Sponsored' in dmap.columns else 'F4E__Sponsored' if 'F4E__Sponsored' in dmap.columns else None
    pop_col = 'Total_Population' if 'Total_Population' in dmap.columns else None
    incharge_col = 'Tsm_Incharge' if 'Tsm_Incharge' in dmap.columns else 'Tsm__Incharge' if 'Tsm__Incharge' in dmap.columns else None
    
    # Add markers for each school
    for idx, row in dmap.iterrows():
        is_selected = selected_school and selected_school in str(row[school_col])
        base_color = division_colors.get(row[division_col], '#6B7280')
        
        # Selected school gets highlighted
        marker_color = '#EF4444' if is_selected else base_color
        radius = 10 if is_selected else 6
        weight = 4 if is_selected else 2
        
        # Create popup content
        popup_html = f"""
        <div style="font-family: 'Segoe UI', sans-serif; min-width: 220px; padding: 5px;">
            <h4 style="margin: 0 0 12px 0; color: {marker_color}; 
                       border-bottom: 2px solid {marker_color}; padding-bottom: 8px;
                       font-size: 14px; font-weight: 600;">
                {row[school_col]}{' ⭐ SELECTED' if is_selected else ''}
            </h4>
            <table style="width: 100%; font-size: 12px; border-collapse: collapse;">
                <tr>
                    <td style="padding: 4px 0; color: #6B7280; font-weight: 500;">Division:</td>
                    <td style="padding: 4px 0; text-align: right; font-weight: 600;">{row[division_col]}</td>
                </tr>
        """
        
        if f4e_col and pd.notna(row.get(f4e_col)):
            popup_html += f"""
                <tr style="background: #F3F4F6;">
                    <td style="padding: 4px 0; color: #6B7280; font-weight: 500;">F4E Sponsored:</td>
                    <td style="padding: 4px 0; text-align: right; font-weight: 600; color: #059669;">{row[f4e_col]}</td>
                </tr>
            """
        
        if pop_col and pd.notna(row.get(pop_col)) and row.get(pop_col) != 0:
            popup_html += f"""
                <tr>
                    <td style="padding: 4px 0; color: #6B7280; font-weight: 500;">Total Population:</td>
                    <td style="padding: 4px 0; text-align: right; font-weight: 600;">{row[pop_col]}</td>
                </tr>
            """
        
        if incharge_col and pd.notna(row.get(incharge_col)) and str(row.get(incharge_col)).strip():
            popup_html += f"""
                <tr style="background: #F3F4F6;">
                    <td style="padding: 4px 0; color: #6B7280; font-weight: 500;">Incharge:</td>
                    <td style="padding: 4px 0; text-align: right; font-weight: 600;">{row[incharge_col]}</td>
                </tr>
            """
        
        popup_html += """
            </table>
        </div>
        """
        
        # Create circle marker
        circle = folium.CircleMarker(
            location=[row[lat_col], row[lon_col]],
            radius=radius,
            popup=folium.Popup(popup_html, max_width=300),
            tooltip=folium.Tooltip(
                f"<b>{row[school_col]}</b><br>Click for details",
                style="font-size: 12px; padding: 8px;"
            ),
            color=marker_color,
            fill=True,
            fillColor=marker_color,
            fillOpacity=0.7 if not is_selected else 0.9,
            weight=weight,
        )
        
        circle.add_to(marker_cluster)
        circle.add_to(divisions_group)
        
        # Add school name label with smart positioning
        # Alternate label positions to avoid overlap
        offset_x = 12 if idx % 2 == 0 else -12
        anchor_x = 'left' if idx % 2 == 0 else 'right'
        
        name_label = folium.DivIcon(
            html=f"""
            <div style="
                font-size: 10px;
                color: {'#EF4444' if is_selected else '#1F2937'};
                font-weight: {'700' if is_selected else '600'};
                white-space: nowrap;
                text-shadow: 1px 1px 2px rgba(255,255,255,0.9), -1px -1px 2px rgba(255,255,255,0.9);
                font-family: 'Segoe UI', sans-serif;
                letter-spacing: 0.3px;
                padding: 2px 6px;
                background: rgba(255,255,255,0.7);
                border-radius: 4px;
                border: {'2px solid #EF4444' if is_selected else '1px solid transparent'};
            ">
                {row[school_col]}{' ★' if is_selected else ''}
            </div>
            """,
            icon_size=(120, 20),
            icon_anchor=(0, 10),
        )
        
        folium.Marker(
            location=[row[lat_col], row[lon_col]],
            icon=name_label,
            interactive=False
        ).add_to(names_group)
    
    # Add coverage area - big rectangle
    folium.Rectangle(
        bounds=[[min_lat, min_lon], [max_lat, max_lon]],
        color='#EF4444',
        fill=True,
        fillColor='#EF4444',
        fillOpacity=0.05,
        weight=2,
        dash_array='5, 10',
        popup='Kitchen Coverage Area'
    ).add_to(coverage_group)
    
    # Add Ikinu HQ marker
    hq_data = dmap[dmap[school_col].str.contains('Ikinu', case=False, na=False)]
    if not hq_data.empty:
        hq = hq_data.iloc[0]
        
        # HQ rectangle
        folium.Rectangle(
            bounds=[
                [hq[lat_col] - 0.008, hq[lon_col] - 0.008], 
                [hq[lat_col] + 0.008, hq[lon_col] + 0.008]
            ],
            color='#DC2626',
            fill=True,
            fillColor='#DC2626',
            fillOpacity=0.1,
            weight=3,
            popup='Ikinu Primary (HQ)'
        ).add_to(coverage_group)
        
        # HQ star marker
        folium.Marker(
            location=[hq[lat_col], hq[lon_col]],
            popup=folium.Popup('<b> HQ: Ikinu Primary</b>', max_width=200),
            tooltip='Headquarters',
            icon=folium.DivIcon(
                html='<div style="font-size: 24px; color: #DC2626; text-shadow: 2px 2px 4px rgba(0,0,0,0.3);">★</div>',
                icon_size=(30, 30),
                icon_anchor=(15, 15),
            )
        ).add_to(coverage_group)
    
    # Add all groups to map
    divisions_group.add_to(m)
    names_group.add_to(m)
    coverage_group.add_to(m)
    
    # Add layer control
    folium.LayerControl(position='topright', collapsed=False).add_to(m)
    
    # Add minimap
    minimap = MiniMap(
        tile_layer='CartoDB positron',
        position='bottomright',
        width=150,
        height=150,
        collapsed_width=25,
        collapsed_height=25
    )
    minimap.add_to(m)
    
    # Fit bounds
    m.fit_bounds([[min_lat, min_lon], [max_lat, max_lon]])
    
    # Custom CSS for better appearance
    custom_css = """
    <style>
        .leaflet-container {
            font-family: 'Segoe UI', sans-serif;
        }
        .leaflet-popup-content-wrapper {
            border-radius: 8px;
            box-shadow: 0 4px 6px -1px rgba(0, 0, 0, 0.1);
        }
        .leaflet-popup-content {
            margin: 12px;
        }
    </style>
    """
    m.get_root().html.add_child(folium.Element(custom_css))
    
    return m


def get_map_html(selected_school=None):
    """Generate map and return as HTML string"""
    m = generate_folium_map(selected_school)
    return m._repr_html_()


# Test standalone

m = generate_folium_map()
m.save('schools_map.html')
print("Map saved to schools_map.html")
m
```
